#### 均值归一化  /  标准化

```python
# 均值归一化，是数据预处理的一种。将样本数据归化成均值为0，方差为1的分布
# 对于样本 X
# for i in feature_size:
#     X[:,i] = X[:,i] - np.mean(X[:,i]) / np.std(X[:,i])
```

```python
# 用归一化则会在不同维度上对数据进行不同的伸缩变化（归一区间，会改变数据的原始距离，分布，信息），
# 使得其呈类圆形。虽然这样样本会失去原始的信息，但这防止了归一化前直接对原始数据进行梯度下降类似的
# 优化算法时最终解被数值大的特征所主导。归一化之后，
# 各个特征对目标函数的影响权重是一致的。这样的好处是在提高迭代求解的精度。
```

```python
# 神经网络、KNN、SVM、PCA等必须对样本数据进行归一化处理
```

#### 协方差

```python
# cov(x_1, x_2) = E[(x_1 - u_1)(x_2 - u_2)] = 1/m * ∑(x_1- u_1)(x_2 - u_2)
# u_1 是指变量x_1的均值
# u_2是指变量x_2的均值
```

```python
# 协方差衡量两个变量x_1，x_2之间的相关性，简单理解就是协方差越大，两个变量的同向变换就越大，否则越小
```

```python
# 相关系数ρ
# ρ = cov(x_1, x_2) / (σ(x_1) * σ(x_2))
# ρ 也可以看出是一种特殊的协方差，是一种剔除了两个变量量纲影响、标准化后的特殊协方差。
# 由于它是标准化后的协方差，因此更重要的特性来了：它消除了两个变量变化幅度的影响，
# 而只是单纯反应两个变量每单位变化时的相似程度。
```

#### 协方差矩阵

<img src = "./图片/协方差矩阵.png">

```python
# 如果将X去均值化 u_i = 0，那么 C = 1/m * X^T*X    
# X = [x_1, x_2, ..., x_n] 是1*n的矩阵，行向量， x_1，x_2 ... 是x的每一项特征
```

#### PCA与样本的协方差矩阵

```python
# 对于某一个样本 m*n 的X而言，我们希望找到一个 n*k 的基矩阵P，使得X*P = Y之后，即对原样本矩阵X做基转换成一个新的样本
# 这个新的样本矩阵Y是m * k ，Y的维度是K维了，这就是降维处理，但是如何找到这个P使得Y尽量不丢失X的信息呢？根据信息论观点，
# Y的样本越随机，信息量越大，也即是样本方差（衡量样本间差异）越大越好，信息量也就越大。
# 所以自然而言就想到尽量让X的各个特征之间越不相关越好(两个特征的协方差为0)。越不相关的特征之间携带的信息量也就越大。
# PCA算法的优化目标就是:    
#    ① 降维后同一纬度的方差最大
#    ② 不同维度之间的相关性为0
# 同一元素的协方差就表示该元素的方差，不同元素之间的协方差就表示它们的相关性。
# 所以PCA的问题也就转换成获得降维后的Y的理想协方差，使得信息量最大。

# Y = 1/m * (X*P)^T * (X*P)  
# 观察Y的协方差矩阵，对角线上的元素是方差、非对角线上的元素是协方差，那么我们的目标就是让Y的协方差矩阵对角化()
# 也就是使得同样的这些样本在那组基上可以达到方差最大，协方差最小。

# Y = P^T * (1/m * X^T * X) * P  =  P^T * C * P 其中C是原来样本矩阵 X的协方差矩阵.
# 因为 P^T * P = I, 设P是正交向量组成的矩阵。目标就是求P
# 根据拉格朗日法 f(P) = max(Y) + λ*(P^T * P - I)  对P求导 C * P = (-λ) * P 
# 所以当P取C的特征向量时，Y能保持最多的信息。

# 如果存在可逆矩阵 A ，使M为对角矩阵，则矩阵A的列等于矩阵的特征向量， M的主对角元素为矩阵M的特征值。

# 让每个协方差元素为0，特征之间不相关，特征不相关不就是说明包含的信息量越大。
# 根据定理 可逆矩阵A就是我们要求的cov的特征向量

# 对角化之后，根据特征值的大小排序，找出前k个最大特征值（绝对值）对应的特征向量，组成基矩阵P。
# 特征值越大，方差也就越大。

# 根据推导 X的协方差矩阵的特征向量组成的基矩阵P，能够获得最大的信息量。 根据定理，实对称阵的特征向量是正交的。 
# 协方差矩阵(X^T * X )就是对称阵。
```

#### PCA的一般过程

```python
# 1、取均值化
# 2、求协方差矩阵
# 3、协方差矩阵对角化
# 4、取前k个最大的特征值对应的特征向量组成基矩阵P n*k 
# 5、X * P 就是PCA降维后的新矩阵
```

#### 奇异值分解 SVD

```python
"""
奇异值分解(Singular Value Decomposition，以下简称SVD)是在机器学习领域广泛应用的算法，
它不光可以用于降维算法中的特征分解（信息提取），还可以用于推荐系统，以及自然语言处理等领域。
是很多机器学习算法的基石。本文就对SVD的原理做一个总结，并讨论在在PCA降维算法中是如何运用运用SVD的。
""" 
```

#### SVD与PCA

```python
"""
奇异值分解能够简约数据，去除噪声和冗余数据。其实它说白了也是一种降维方法，
将数据映射到低维空间。看到这里其实就会想，它和主成分分析（PCA）有什么联系或者差异呢？
奇异值分解和主成分分析一样，也是告诉我们数据中重要特征，

奇异值是数据矩阵的转置乘以该矩阵的特征值的平方根（Data^T * Data特征值的平方根）。


PCA降维是 A*A^T 的前n个特征值对应的特征向量方向的投影，也就是V矩阵
但是SVD 不仅有 A*A^T 投影还有 A^T * A 的特征值对应的特征方向ui的投影，也即是U矩阵。
所以PCA只有一个主成分方向，而SVD有两个方向，分别是U和V

"""
```

<img src= "./图片/奇异值分解.jpg">

```python
"""
SVD就是矩阵分解。
∑矩阵的对角元素是从大到小排列的。这些对角元素就称为奇异值。
"""
```

#### 怎么选取奇异值

```python
"""
在某个奇异值的数目r之后，其他的奇异值均置0，也就是我们仅保留r个重要特征，其余特征都是噪声或者冗余特征。
那么问题来了，这个r到底是多少勒？如何选取呢？

确定要保留的奇异值个数有很多启发式的策略，其中一个典型的做法就是保留矩阵90%的能量信息。
为了计算能量信息，将所有的奇异值求平均和，直到累加到总值的90%为止。

另一个启发式策略是当矩阵有上万个奇异值时，保留前面的2000个或3000个。
"""
```

#### SVD 证明

```python
"""
1、首先看特征值和特征向量
A为对称的n阶方阵，那么有   A = W * Σ * W^-1 
W是A的特征向量张成的空间，一般取W的每个向量为单位向量，这样 W^T * W=I
所以 A = W * Σ * W^T

因为对称阵的特征向量两两正交，它的所有特征向量能够组成一个正交基，如果特征向量单位化，那么就是单位正交基。


问题来了，如果A不是方阵呢？ m * n维呢？ SVD分解

A = U * Σ * V

设：
(A^T * A)*vi = λi * vi
V^T * V = I 

Σ 是对角矩阵，主对角线上的值均为非负值，且按递减顺序排列。


U^T * U = I
Σ^T * Σ = Σ^2

A = U * Σ * V^T 
A^T = V * Σ^T * U^T
那么有
T = A^T * A = V * Σ^T * U^T * U * Σ * V^T = V * Σ^2 V^T
T * V = V * Σ^2

同理：
D = A * A^T = U * Σ * V^T * V * Σ^T * U^T = U * Σ^2 U^T
D * U = U * Σ^2

就是说 矩阵U是A*A^T的特征向量，矩阵V是A^T*A的特征向量。
并且  A*A^T 和 A^T*A 特征值相同。
通常情况下可以认为矩阵的SVD是唯一的，但当存在多个相同奇异值时，矩阵可以有不同的SVD分解。


如果把A看出下面的形式

A =  λ1 *  u1 * v1^T + λ2 *  u2 * v2^T +  ... + λr *  ur * vr^T   r为最后一个奇异值
ui就是U的第i列列向量
vi就是V的第i列列向量

"""
```

#### SVD性质

```python
"""
A m×n = Um×m * Σm×n * V^Tn×n ≈ Um×k * Σk×k * V^Tk×n

这里的k就是奇异值的个数，也就是说，我们其实不需要那么多的零向量。
"""
```

#### SVD的两种解释

```python
"""
1、文档-词矩阵分解成文档-主题矩阵 * 奇异值矩阵 * 主题-词矩阵  这就是 LSA ，∑中的奇异值就是主题。矩阵分解和概率模型联系起来了。

2、矩阵U或V的列向量是矩阵A在空间中的最佳标准正交基（最小均方根误差RMSE，等价于投影方差最大化）。
例如在此例中，如果把每个文档看成是五维向量，那么V矩阵的第一个列向量所代表的方向就是所有文档最大化投影方差的方向（即数据信息最大化）。
∑矩阵的对应值可以看成是投影方差的一种度量。A * V= ∑ * U 对应着所有文档在这组标准正交基上的投影坐标。

"""
```

#### U 和 V

```python
"""
A * V= ∑ * U，啥意思？ 是不是就是对A在主成分方向进行投影。
左奇异矩阵U可以用于行数的压缩（样本压缩，可以理解为是样本归 类）。


相对的，右奇异矩阵V可以用于A的列数即特征维度的压缩，也就是我们的PCA降维。

ui和vi受制于σi
"""
```

#### SVD应用 -- 用户推荐

```python
# V = A^T * U * ∑^-1  V 就是代表样本映射后的样本矩阵，就是降维之后的样本矩阵
# 来一个新用户A_new ,计算出V，然后利用余弦距离看V跟之前哪个用户更相似。
# 找到这个用户之后，就将该用户喜欢的东西推荐给A_new
```

#### 梯度上升法

```python
#  目标函数
def f(W, X):
    return np.sum((X.dot(W)**2)) / len(X)
```

```python
# 通用的求梯度公式 J = 1/m * ∑(X*w)^2 
# ▽J = [dJ/dw_1, dJ/dw_2, ...,  dJ/dw_n]^T
# 梯度▽J 是n*1的列向量。
# 用向量化表示 ▽J = 1/m * 2* X^T * (X*W)
def df_math(W, X):
    return X.T.dot(X.dot(W)) * 2 / len(X)
```

```python
# 验证梯度公式
# 注意W是单位方向向量
def df_debug(W, X, epsilon=0.0001):
    res = np.empty(len(W))
    for i in range(len(W)):
        w_1 = W.copy()
        # 梯度增加一个小值
        w_1[i] += epsilon
        w_2 = W.copy()
        # 梯度减少一个小值
        w_2[i] -= epsilon
        res[i] = (f(w_1, X) - f(w_2, X)) / (2 * epsilon)
    return res
```

```python
# W 是映射后的轴的方向向量，所有W中的每一个分量都是单位向量
# np.linalg.norm(W) 求W的模
def direction(W):
    return W / np.linalg.norm(W)

def gradient_ascent(df, X, initial_W, eta, n_iters = 1e4, epsilon=1e-8):
    W = direction(initial_W)
    cur_iter = 0
    
    while cur_iter < n_iters:
        gradient = df(W, X)
        last_W = W
        W = W + eta *gradient
        # 每次都要使得W变成单位向量，这是用梯度上升法求PCA要注意的问题。
        W = direction(W)
        if (abs(f(W, X) - f(last_W, X)) < epsilon):
            break
        cur_iter += 1
    
    return W
```

```python
# 注意初始的theta不能为0，为啥因为df_math(0, X) 恒为0，0本身就是一个极值点了
# 所以intial_theta 随机就好
initial_W = np.random.randn(X.shape[1])
```

```python
eta = 0.001
```

```python
# PCA问题不能使用standard scaller 归一化
# 其实我们的demean步骤就已经做了归一化的第一步，均值化，只是没有除以标准差。
```

```python
# 求得的结果就是映射到这个（结果表示的）轴后，方差是最大的。
gradient_ascent(df_debug, X_demean, initial_W, eta)
```

```python
W = gradient_ascent(df_math, X_demean, initial_W, eta)
W
```

#### PCA的其他主成分

```python
# 怎么求PCA的主成分
# 首先求得了第一个主成分，然后第二个主成分就是去掉这个主成分的分量就好。
# 怎么去掉呢？
# X - X_proj_1即可，几何意义就是X_proj_1的垂直方向，仔细想一想这个不就是取X_proj_1的不相关的其他分量吗。 bingo，其实也跟这个
# 协方差矩阵联系起来了
```

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
X = np.empty((100,2))
X[:,0] = np.random.uniform(0., 100, size=100)
X[:,1] = 0.75 * X[:,0] + 3. + np.random.normal(0,10,size=100)
```

```python
plt.scatter(X[:,0], X[:,1])
plt.show()
```

```python
# df就是df_math
def df(W, X):
    return X.T.dot(X.dot(W)) * 2 / len(X)
```

```python
# 1、求第一主成分，其实也就是第一个映射轴方向
# 就是梯度上升法
def first_component(X, initial_W, eta, n_iters=1e4, epsilon=1e-8):
    W = direction(initial_W)
    cur_iter = 0
    
    while cur_iter < n_iters:
        gradient = df(W, X)
        last_W = gradient
        W += gradient*eta
        W = direction(W)
        if abs(f(W, X) - f(last_W, X)) < epsilon:
            break
        cur_iter += 1
    return W
```

```python
initial_W = np.random.random(X.shape[1])
eta = 0.01
W = first_component(X, initial_W, eta)
W
```

```python
# 求第二个映射轴， 垂直于第一个轴
X2 = np.empty(X.shape)
# for i in range(len(X)):
#     X2[i]  = X[i] - X[i].dot(W) * W   # X[i].dot(W) 是X在W上映射的模，再乘以单位向量W，就是W的分量 等于X_proj_1

# 向量化
X2 = X - X.dot(W).reshape(-1,1)* W
```

```python
# 计算一下在X2的方向
W2 = first_component(X2, initial_W, eta)
W2
```

```python
# W * W2应该为0
W.dot(W2)
```

```python
# 求前N个主成分,封装。
def first_n_components(n, X, eta=0.01, n_iters=1e4, epsilon=1e-8):
    X_pca = X.copy()
    # 首先去均值化
    X_pca = demean(X)
    
    res = []
    for i in range(n):
        initial_W = np.random.random(X_pca.shape[1])
        W = first_component(X_pca, initial_W, eta)
        res.append(W)
        
        X_pca = X_pca - X_pca.dot(W).reshape(-1,1) * W
    return res
```

#### 高维数据向低维数据映射

```python
# 数据样本集X包含m个样本，m*n维数据，
# 映射之后的样本集W， k*n维。k表示前k个主成分，每个主成分有n个元素。

# X*M^T  就是m*k维数据，从n维压缩到了k维 得到低维数据X'
# 同样低维数据 X’* M  就是m*n维
```

```python
class PCA(object):
    def __init__(self,n_components):
        """初始化PCA"""
        assert n_components >=1, "n_components must > 1"
        self.n_components = n_components
        # components_ 每个主成分是谁
        self.components_ = None
        
    def fit(self, X, eta=0.01, n_iters=1e4):
        """求前n个主成分"""
        assert self.n_components <= X.shape[1], "n_components must not be greater than n features"
        def demean(X):
            return X - np.mean(X, axis=0)
        def f(W, X):
            return np.sum((X.dot(W)**2)) / len(X)
        def df(W, X):
            return X.T.dot(X.dot(W)) * 2. / len(X)
        
        # 方向向量
        def direction(W):
            return W / np.linalg.norm(W)

        def first_component(X, initial_W, eta, n_iters=1e4, epsilon=1e-8):
            W = direction(initial_W)
            cur_iter = 0

            while cur_iter < n_iters:
                gradient = df(W, X)
                last_W = gradient
                W += gradient*eta
                W = direction(W)
                if abs(f(W, X) - f(last_W, X)) < epsilon:
                    break
                cur_iter += 1
            return W

        # 首先去均值化
        X_pca = demean(X)
        self.components_ = np.empty(shape=(self.n_components, X.shape[1]))
        res = []
        for i in range(self.n_components):
            initial_W = np.random.random(X_pca.shape[1])
            W = first_component(X_pca, initial_W, eta)
            # components_ 是k*n的矩阵
            self.components_[i,:] = W

        X_pca = X_pca - X_pca.dot(W).reshape(-1,1) * W
        return self

    def transform(self,X):
        """将给定的X，映射到各个主成分分量"""
        assert X.shape[1] == self.components_.shape[1]
        return X.dot(self.components_.T)
    
    def inverse_transform(self, X):
        """将映射后的低维数据反映射会原来的特征空间"""
        assert X.shape[1] == self.components_.shape[0]
        return X.dot(self.components_)
        
    def __repr__(self):
        return "PCA(n_components = %d)" % self.n_components
```

#### 从高维数据到低维数据

```python
pca = PCA(n_components=1)
pca.fit(X)
```

```python
# 降维后的结果
X_reduction = pca.transform(X)
X_reduction.shape
```

```python
# 反向转回到高维
X_restore = pca.inverse_transform(X_reduction)
X_restore.shape
```

```python
plt.scatter(X[:,0], X[:,1])
plt.scatter(X_restore[:,0], X_restore[:,1], color='r')
plt.show()
```

#### 使用PCA降噪

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
from sklearn import datasets

digits = datasets.load_digits()
X = digits.data
y = digits.target
```

```python
# 首先给数字加上噪音
noisy_digits = X + np.random.normal(0,4,size=X.shape)
noisy_digits.shape
```

```python
# 然后取出一些例子
# 0
example_digits = noisy_digits[y==0,:][:10]
for num in range(1,10):
    # 1-9
    X_num = noisy_digits[y==num,:][:10]
    # 将这些例子垂直累加
    example_digits = np.vstack([example_digits, X_num])
```

```python
example_digits.shape
```

```python
def plot_digits(data):
    fig, axes = plt.subplots(10,10,figsize=(10,10), subplot_kw={'xticks':[], 'yticks':[]}, gridspec_kw = dict(hspace=0.1, wspace=0.1))
    for i,ax in enumerate(axes.flat):
        ax.imshow(data[i].reshape(8,8), cmap='binary',interpolation='nearest', clim=(0,16))
```

```python
plot_digits(example_digits)
# 看出噪声还是蛮大。
```

```python
# 降噪处理
pca = PCA(0.5)
pca.fit(noisy_digits)
```

```python
pca.n_components_
# 12维
```

```python
# 将样本映射到低维
components = pca.transform(example_digits)
# 再将低维反映射到高维
filter_digits = pca.inverse_transform(components)
```

```python
# 再打印看看，已经清晰很多了
plot_digits(filter_digits)
```

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
random_indexes = np.random.permutation(len(faces.data))
X = faces.data[random_indexes]
```

```python
example_faces = X[:36, :]
example_faces.shape
```

```python
def plot_faces(faces):
    fig, axes = plt.subplots(6,6, figsize=(10,10), subplot_kw={'xticks':[], 'yticks':[]}, gridspec_kw=dict(hspace=0.1,wspace=0.1))
    for i,ax in enumerate(axes.flat):
        ax.imshow(faces[i].reshape(62,47), cmap='bone')
    plt.show()
```

```python
# 打印前36个重要的特征脸
plot_faces(pca.components_[36:])
```
