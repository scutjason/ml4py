```python
import matplotlib.pyplot as plt
```

```python
import numpy as np
# 1-10，切分100
X = np.linspace(0,10,100)
X
```

```python
y = np.sin(X)
y
```

```python
plt.plot(X,y)
```

```python
cosy= np.cos(X)
```

```python
cosy.shape
```

```python
siny = y.copy()
```

```python
# 绘制多条曲线
plt.plot(X, siny)
plt.plot(X, cosy)
plt.show()
```

```python
# 添加颜色
plt.plot(X, siny)
plt.plot(X, cosy, color='red')
plt.show()
```

```python
# 添加样式
plt.plot(X, siny)
plt.plot(X, cosy, color='red', linestyle = '--')
plt.show()
```

```python
# 调节坐标轴
# x -1到11
# y -2到2
plt.plot(X, siny)
plt.plot(X, cosy, color='red', linestyle = '--')
plt.axis([-1,11, -2,2])
plt.show()
```

```python
# 坐标轴标签
plt.plot(X, siny)
plt.plot(X, cosy, color='red', linestyle = '--')
plt.xlabel("x axis")
plt.ylabel("y value")
plt.show()
```

```python
# 坐标轴图示
plt.plot(X, siny, label="sin(x)")
plt.plot(X, cosy, color='red', linestyle = '--', label="cos(x)")
plt.xlabel("x axis")
plt.ylabel("y value")
plt.legend()
plt.show()
```

```python
# 添加标题
plt.plot(X, siny, label="sin(x)")
plt.plot(X, cosy, color='red', linestyle = '--', label="cos(x)")
plt.xlabel("x axis")
plt.ylabel("y value")
plt.legend()
plt.title("Welcom to the ML World!")
plt.show()
```

### 读取数据

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
from sklearn import datasets
```

```python
iris= datasets.load_iris()
```

```python
iris.keys()
```

```python
print(iris.DESCR)
```

```python
iris.data
```

```python
iris.data.shape
```

```python
# 特征名字
iris.feature_names
```

```python
# 每个鸢尾花样本对应的类型
iris.target
```

```python
# 种类
iris.target_names
```

```python
# 取data的第0、1列
X = iris.data[:, :2]
```

```python
X.shape
```

```python
# X的第0列在x轴，第1列在y轴
plt.scatter(X[:,0], X[:,1])
plt.show()
```

```python
y = iris.target
y
```

```python
# 选取y==0的行的第0列，
plt.scatter(X[y==0, 0], X[y==0,1], color="red") # 挑选出target=0的点
plt.scatter(X[y==1, 0], X[y==1,1], color="blue")
plt.scatter(X[y==2, 0], X[y==2,1], color="green")
plt.show()
```

```python
# 设置点的样式
plt.scatter(X[y==0, 0], X[y==0,1], color="red", marker="o")
plt.scatter(X[y==1, 0], X[y==1,1], color="blue", marker="x")
plt.scatter(X[y==2, 0], X[y==2,1], color="green", marker="+")
plt.show()
```
